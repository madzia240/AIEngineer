# 2. Seaborn

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics.

<img src="https://drive.google.com/uc?export=view&id=1iYWtn_YCfkk30BHSvPHgKZi6UUN0kPwA" alt="Seaborn: tabular data" title="Seaborn: logo" align="center" width="500px" hspace="30px" vspace="10px"/>

Here are some key points about Seaborn:
* **<font color='##13e5e1'>simplified syntax</font>**: optimized functions and default settings for common plot types, it allows you to create visually appealing plots with minimal code
* **<font color='##13e5e1'>statistical visualizations</font>**: this library is particularly useful for exploring relationships between variables, visualizing distributions, and highlighting patterns in data. Additionaly, it extends Matplotlib's capabilities by providing additional statistical functionalities. It includes features such as regression plot, correlation heatmap, categorical plots, time series visualization, and support for statistical annotations.
* **<font color='##13e5e1'>Integration with Pandas</font>**: it seamlessly integrates with the popular data manipulation library Pandas, it can directly accept Pandas DataFrames as input, making it convenient for working with tabular data
* **<font color='##13e5e1'>Attractive default styles</font>**: it comes with aesthetically pleasing default styles and color palettes

Overall, Seaborn is a powerful library for data visualization, particularly in the realm of statistical graphics. It complements Matplotlib and Pandas, making it easier to create beautiful and informative visualizations with minimal effort.

## Scatter plot with regression model

The biggest advantage of using Seaborn is the variety of functions for creating graphs. For example, **<font color='##13e5e1'>creating a scatter plot combined with a regression model</font>** is just one line of code using the `sns.lmplot()` function.

This can be done in two ways:
* The first (recommended) way is to pass a DataFrame to the data argument while passing column names to the x and y arguments.
* The second way is to pass the data series directly to the axis arguments. - **<font color='red'>deprecated</font>**

```python
sns.lmplot(x='Alcohol', y='Magnesium', data=wine)
plt.show()
```

In fact, Seaborn does not have a dedicated scatter plot function. To create this graph, we use the Seaborn function `sns.lmplot()` to fit and plot the regression line.

Fortunately, each plot function has some useful options that can be set. Here's how we can adjust the lmplot() function to your needs:
* set `fit_reg = False` to remove the regression line, since we want a scatter plot
* set `hue = 'Class'` to color our points according to the class of the example.

```python
sns.lmplot(
    x='Alcohol',
    y='Magnesium',
    data=wine,
    fit_reg=False,
    hue='Class'
)

plt.show()
```

## Violin plot

Another advantage of Seaborn is that it has **<font color='##13e5e1'>predefined styles</font>**. The default one is "darkgrid."

Here we have changed the style from the default `"whitegrid"` to `"darkgrid"` when creating a violin chart. Such a graph shows the distribution of values in a similiar but slightly different way than a boxplot.

```python
sns.set_style('darkgrid')

sns.violinplot(x='Class', y='Alcohol', data=wine, hue="Class")
plt.show()
```

Violin charts are great for visualizing distribution. However, because there are only 170 wines in our dataset, we can simply display each point - with swarmplot showing the position of each point and stacking those with similar values.

```python
sns.swarmplot(x='Class', y='Alcohol', data=wine, hue='Class')
plt.show()
```

Combine the best of several types of charts and libraries:
* First, enlarge our figure with Matplotlib.
* Next, plot a violinplot with the parameter `inner=None`.
* Inside the violinplot, draw a swarmplot with black dots.
* Finally, set the title using Matplotlib.

```python
sns.set_style('whitegrid')

sns.violinplot(x='Class', y='Alcohol', data=wine, inner=None, hue="Class", alpha=0.5)
sns.swarmplot(x='Class', y='Alcohol', data=wine, hue="Class", alpha=1)

plt.ylim([10.5, 15.5])
plt.title('Alcohol according to the type of wine')
plt.show()
```

## Histogram with KDE

In Seaborn, `displot()` is a function used to create a variety of plots for univariate distributions. It supports various types of plots, including histograms, **<font color='##13e5e1'>kernel density estimation (KDE) plots</font>**, and rug plots. When using `displot()` with the kind parameter set to 'kde', it generates a KDE plot.

A KDE plot visualizes the **<font color='##13e5e1'>estimated probability density function of a continuous variable</font>**. It provides a smooth curve that approximates the underlying distribution of the data.

```python
sns.displot(wine.Alcohol, kde=True)
plt.show()
```

## Countplot

A countplot is a **<font color='##13e5e1'>type of bar plot in which the count or frequency of each category</font>** in a categorical variable is represented by the height of the bars. Seaborn provides a convenient function called `sns.countplot()` to create countplots.



```python
sns.countplot(x='Class', data=wine)
plt.show()
```

## KDE & joint plots for 2D data

To create a 2D kernel density estimation (KDE) plot in Seaborn, you can use the `sns.kdeplot()` function. This plot allows you to visualize the density of a bivariate dataset, representing the joint distribution of two continuous variables.

```python
sns.kdeplot(
    data=wine,
    x='Alcohol',
    y='Magnesium'
)
plt.show()
```

```python
sns.jointplot(
    data=wine,
    x='Alcohol',
    y='Magnesium'
)
plt.show()
```
